```python
import pandas as pd
```

```python
df=pd.read_excel("missed_data.xls")
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({'temperature':0,'windspeed':0,'event':'no event'})
new_df
```

```python
##carry forwards the values from previous value
new_df=df.fillna(method="ffill")
new_df
```

```python
##carrybacks value
new_df=df.fillna(method="bfill")
new_df
```

```python
new_df=df.fillna(method="ffill",limit=1)
new_df
```

```python
##replacing wth intermediate values
new_df=df.interpolate()
new_df
```

```python
##in contraint with time
new_df=df.interpolate(method="time")
new_df
```

```python
##dropping na values
new_df=df.dropna()
new_df
```

```python
##row having complete nan values will be dropped
new_df=df.dropna(how="all")
new_df
```

```python
new_df=df.dropna(thresh=1)
new_df
```

```python
new_df=df.fillna({'event':'no_event'})
new_df
```
